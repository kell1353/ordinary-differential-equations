# Analyzing results

We will use python and this jupyter notebook to plot and analyze the results obtained from the
fortran source code.

## Plotting results

For this we will use the `numpy` and `matplotlib.pyplot` packages

```python
# Importing numpy and matplotlib
import numpy as np 
import matplotlib.pyplot as plt
import f90nml
```

```python
# different styles are predefined that give different appearance to the plots. This one emulates the style of 
# the figures you'll find on statisical analysis news site fivethirtyeight.com
#plt.style.use('fivethirtyeight')
plt.style.use('seaborn')
```

# Planetary Motion

```python
# reading data stored in planetary_motion.dat

file_name = 'planetary_motion.dat'

# loadtxt reads all the data in a file and stores them in an array (assuming it's all floats)
# we use skiprows=8 to avoid the header
all_data = np.loadtxt(file_name,skiprows=8)

# We transpose the array in order to be able to extract each column individually
all_data = all_data.transpose()

time = all_data[0]
x1_points = all_data[1]
y1_points = all_data[2]
x2_points = all_data[3]
y2_points = all_data[4]
energy = all_data[5]
angular_momentum = all_data[6]
```

### Initial Conditions of the System

```python
# Read the values from the namelist file
parameters = f90nml.read('planetary_motion.namelist')

initial_pos_1 = parameters['initial_conditions']['initial_pos_1']
initial_pos_2 = parameters['initial_conditions']['initial_pos_2']
initial_vel_1 = parameters['initial_conditions']['initial_vel_1']
initial_vel_2 = parameters['initial_conditions']['initial_vel_2']

final_time = parameters['solution_parameters']['final_time']
n_points = parameters['solution_parameters']['n_steps']

# Show the initial conditions 
print('Final Time: ' + str(final_time))
print('Inital Position of Object 1: ' + str(initial_pos_1))
print('Inital Position of Object 2: ' + str(initial_pos_2))
print('Inital Velocity of Object 1: ' + str(initial_vel_1))
print('Inital Velocity of Object 2: ' + str(initial_vel_2))
```

### Plot the Positions of the Masses

```python
fig, ax = plt.subplots()

# Plotting the two masses
plt.plot(x1_points, y1_points, linewidth = .5, label='First Planet')
plt.plot(x2_points, y2_points, linewidth = .5, label='Second Planet')

# Plotting the central mass
large_mass = plt.Circle((0, 0), 3, color='b')
ax.add_artist(large_mass)

# Plot Settings
plt.xlabel('x')
plt.ylabel('y')
plt.axis('scaled')
plt.show()
```

### Graph of the total energy and angular momentum

```python
figure, (ax1, ax2) = plt.subplots(1, 2)

# Settings for plot 1
ax1.plot(time, energy, linewidth = .75)
ax1.set_title('Energy', size= 12)
ax1.set_xlim([0, time[-1]])
ax1.set_ylim(-2 + min(energy), 2 + max(energy))
ax1.set_xlabel('time', fontsize = 13.0)
ax1.set_ylabel('energy', fontsize = 13.0)

# Settings for plot 2
ax2.plot(time, angular_momentum, linewidth = .75)
ax2.set_title('Angular Momentum', size= 12)
ax2.set_xlim([0, time[-1]])
ax2.set_ylim(-2 + min(angular_momentum), 2 + max(angular_momentum))
ax2.set_xlabel('time', fontsize = 13.0)
ax2.set_ylabel('angular momentum', fontsize = 13.0)

plt.tight_layout()
plt.show()
```
